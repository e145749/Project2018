```python
import pandas as pd
import numpy as np
import seaborn as sns
#import matplotlib as plt
import matplotlib.pyplot as plt
import sklearn.preprocessing as sp
%matplotlib inline
sns.set_style('whitegrid')
sns.set_context("poster")
sns.set_context("talk", 1.0, {"lines.linewidth": 3})
```

```python
df_s = pd.read_csv("./取得データセット/受信機1/direction.csv")
df_s.columns = ["UUIDs","Major","Minor","Accuracy","Proximity","TxPower","RSSI","device_name","timestamp","x","y"]
```

```python
df_s
```

```python
x = np.array([1,2,3,4])
dire = ["north","east","south","west"]
y = []
for i in dire:
    y.append(int(df_s[(df_s.x == i)].RSSI.count()))
plt.bar(x,y,tick_label=dire, align="center")
```

```python
x = np.array([1,2,3,4])
dire = ["north","east","south","west"]
y = []
for i in dire:
    y.append(int(df_s[(df_s.x == i)].RSSI.mean()))
plt.bar(x,y,tick_label=dire, align="center")
```
