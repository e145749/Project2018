```python
import pandas as pd
import numpy as np
import seaborn as sns
#import matplotlib as plt
import matplotlib.pyplot as plt
import sklearn.preprocessing as sp
%matplotlib inline
sns.set_style('whitegrid')
sns.set_context("poster")
sns.set_context("talk", 1.0, {"lines.linewidth": 3})
```

```python
#データセット取得
df0 = pd.read_csv("./取得データセット/受信機1/nkmr1-dataset0.csv")

#それぞれのカラム名の指定
df0.columns = ["UUIDs","Major","Minor","Accuracy","Proximity","TxPower","RSSI","device_name","timestamp","X","y"]

#不要なデータの消去
df0 = df0[df0.UUIDs != "00000000e2681001b000001c4d942633"]
```

```python
print("RSSIのデータ数:%s" % df0.RSSI.count())
print("RSSIの平均値:%s" % df0.RSSI.mean())
print("RSSIの分散値:%s\n" % df0.RSSI.std())

ax1 = plt.subplot()
ax1.plot(df0.index,df0.RSSI,"bo")
ax1.set_ylim([-100,-50])
plt.savefig("battery.png")
```
